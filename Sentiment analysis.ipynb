```python
import pandas as pd
import matplotlib.pyplot as plt
import nltk
%matplotlib inline
```

```python
train =  pd.read_csv("C:\\Users\\Naveen\\Desktop\\AV_Hackthon\\Identify the Sentiments\\train.csv")
test = pd.read_csv("C:\\Users\\Naveen\\Desktop\\AV_Hackthon\\Identify the Sentiments\\test.csv")
```

```python
train.head(10)
```

```python
#Cleaning data
import re
train['tweet'] = train['tweet'].apply(lambda x: ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t]) |(\w+:\/\/\S+)", " ", x).split()))
test['tweet'] = test['tweet'].apply(lambda x: ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t]) |(\w+:\/\/\S+)", " ", x).split()))
```

```python
from textblob import TextBlob

def Clasify(tweets):
    predections = []
    for tweet in tweets:    
        analysis = TextBlob(tweet)
        print(analysis.sentiment)
        if analysis.sentiment[0]>0:
            predections.append(0)
        else:
            predections.append(1)
        print("")
    return predections
```

```python
TrainPredections = Clasify (train['tweet'])
```

```python
TestPredections = Clasify (test['tweet'])
```

```python
submission = pd.DataFrame()
submission['id'] = test['id']
submission['label'] = TestPredections
submission.head()
```

```python
submission.to_csv("C:\\Users\\Naveen\\Desktop\\AV_Hackthon\\Identify the Sentiments\\predections.csv",index=False)
```
